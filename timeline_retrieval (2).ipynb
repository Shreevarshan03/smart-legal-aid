```python
df = pd.read_csv('casedocs.csv',header=None)
df.columns = ["Text"]
df
```

```python
z=[]
for i in range(2914):
    z.append(df.Text[i])
```

```python
jud=[]
for c in z:
    for j in range(len(c.split())):
        p=c.split()
        if p[j] =="India" and p[j-2]=="Court":
            q=str(p[j+1])+ " "+str(p[j+2])+" "+str(p[j+3])
            jud.append(q)
            break
```

```python
jud
```

```python
df['Start']=jud
print(df)
```

```python
f=[]
for c in z:
    index=0
    for j in range(len(c.split())):
        index+=1
        if index>=len(c.split()):
            f.append('0')
            continue
        p=c.split()
        if (p[j] =="No." or p[j]=="No"):
            print(p[j+3])
            f.append(p[j+3])
            break
```

```python
df['End']=f
print(df)
```
